```python
text_edited = pd.read_csv('/content/drive/My Drive/reviews_0_250.csv', low_memory=False)

text_edited.info()
```

```python
text_edited['brand_name'] = text_edited['brand_name'].astype(str)
text_edited['review_text'] = text_edited['review_text'].astype(str)
text_edited['review_title'] = text_edited['review_title'].astype(str)
```

```python
import nltk
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
stopwords = stopwords.words('english')
```

```python
textb = text_edited[text_edited['rating'] !=3]
textb['sentiment']= textb['rating'].apply(lambda rating: +1 if rating >3 else -1)
textb.head()
```

```python
def remove_punctuation(text):
    final="".join(u for u in text if u not in ("?",".",";",":","!",'"'))
    return final

textb['review_text'] = textb['review_text'].apply(remove_punctuation)
textb=textb.dropna(subset=['review_title'])
textb['review_title']=textb['review_title'].apply(remove_punctuation)
```

```python
textb=textb.dropna(subset=['review_text'])
```

```python
textb=textb.dropna(subset=['brand_name'])
```

```python
new_text = textb[['review_title','review_text','brand_name','sentiment']]
new_text.head()
```

```python
new_text['review_title_and_text'] = new_text['review_title'] + new_text['review_text'] + new_text['brand_name']
```

```python
new_text = new_text.reset_index(drop=True)
```

```python
new_text['review_title_and_text'] = new_text['review_title_and_text'].str.lower()
```

```python
new_text['review_title_and_text'] = new_text['review_title_and_text'].apply(lambda x: [ _ for _ in x.split()])
```

```python
new_text['review_title_and_text'] = new_text['review_title_and_text'].apply(lambda x: [ _ for _ in x if _ not in stopwords])
```

```python
index = new_text.index
```

```python
index = new_text.index
new_text['random_number'] =np.random.randn(len(index))

train=new_text[new_text['random_number']<=0.8]
test=new_text[new_text['random_number']>0.8]

file_path = '/content/drive/My Drive/test.csv'

test.to_csv(file_path, index=False)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

tokenizer = Tokenizer()
tokenizer.fit_on_texts(new_text['review_title_and_text'])
sequences = tokenizer.texts_to_sequences(new_text['review_title_and_text'])
max_seq_length = max([len(x) for x in sequences])
train_sequences = [sequences[i] for i in train.index]
test_sequences = [sequences[i] for i in test.index]

X_train = pad_sequences(train_sequences, maxlen=max_seq_length)
X_test = pad_sequences(test_sequences, maxlen=max_seq_length)
y_train = train['sentiment'].values
y_test = test['sentiment'].values
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Embedding

model = Sequential()
model.add(Embedding(input_dim=len(tokenizer.word_index) + 1, output_dim=100, input_length=max_seq_length))
model.add(LSTM(128))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X_train , y_train, epochs=1, validation_split=0.2)
```

```python
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {accuracy * 100:.2f}%")
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

def preprocess_text(text, tokenizer, max_seq_length):

    sequences = tokenizer.texts_to_sequences([text])
    padded_sequences = pad_sequences(sequences, maxlen=max_seq_length)
    return padded_sequences
```

```python
def predict_sentiment(processed_text, model):
    prediction = model.predict(processed_text)
    return prediction
```
